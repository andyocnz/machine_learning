```python
def listToString(s):  
    
    # initialize an empty string 
    str1 = ", "  
    
    # traverse in the string   
    for ele in s:  
        str1 += ele   
    
    # return string   
    return str1  
        
        
# Driver code     

c=(listToString(lines))  
```

```python
# instantiate a word cloud object
bc = WordCloud(
    background_color='white',
    max_words=1000,
    stopwords=stopwords,width=800, height=400
)

# generate the word cloud
bc.generate(wantedtext)
```

```python
stopwords.add('Re') # add the words said to stopwords
stopwords.add('Re:')
stopwords.add(' Re')
stopwords.add('RE  ')
stopwords.add('Taranaki')
stopwords.add('Commercial')
stopwords.add('commercial')
stopwords.add('re')
stopwords.add('New')
stopwords.add('sale')
stopwords.add('RE')
stopwords.add('Bartercard')
stopwords.add('Wanted')
stopwords.add('New')
stopwords.add('Bartercard wanted')
stopwords.add('-')
stopwords.add(':')
stopwords.add('&')
stopwords.add('x')
stopwords.add('2')
stopwords.add('3')
stopwords.add('/')
stopwords.add('4')
```

```python
# verify top 50 words
filtered_words = [word for word in wantedtext.split() if word not in stopwords]
counted_words = collections.Counter(filtered_words)

words = []
counts = []
for letter, count in counted_words.most_common(50):
    words.append(letter)
    counts.append(count)
                  
                  
colors = cm.rainbow(np.linspace(0, 1, 10))
rcParams['figure.figsize'] = 20, 10

plt.title('Top words in the headlines vs their count')
plt.xlabel('Count')
plt.ylabel('Words')
plt.barh(words, counts, color=colors)                  
```
